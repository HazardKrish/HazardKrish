```python
#using matplotlib library in python on google colab

import matplotlib.pyplot as plt
```

```python
x=[1,3,5,7,9]
y1=[50,40,70,80,20]
y2=[80,20,20,50,60]
y3=[50,30,70,60,10]
y4=[20,40,60,50,70]

plt.plot(x,y1,'g',label='Enfield', linewidth=2)
plt.plot(x,y2,'c',label='Bajaj',linewidth=1)
plt.plot(x,y3,'k',label='Yahama',linewidth=3)
plt.plot(x,y4,'y',label='Harley Davidson',linewidth=4)

plt.title('Bike details inline plot')
plt.ylabel('Distance in kms')
plt.xlabel('Days')

plt.legend()
```

```python
x1=[0.25,1.25,2.25,3.25,4.25]
y1=[50,40,80,30,20]

plt.bar(x1,y1,label="BMW",color='k',width=0.4)

x2=[0.75,1.75,2.75,3.75,4.75]
y2=[30,20,20,50,60]
plt.bar(x2,y2,label="Audi",color='c',width=0.3)

plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Information')

plt.legend()
```

```python
x=[5,6,3,7,2]
y=["A","B","C","D","E"]

plt.barh(y,x,color="orange")

plt.show()
```

```python
x=[5,4,2,8,9]
y=[10,5,4,8,2]

plt.scatter(x,y)

plt.show()
```

```python
#multiple scatter plot

x1=[1,1.5,2,2.5,3,3.5,3.6]
y1=[7.5,8,8.5,9,9.5,10,10.5]

x2=[8,8.5,9,9.5,10,10.5,11]
y2=[3,3.5,3.7,4,4.5,5,5.2]

plt.scatter(x1,y1,label="High income low salary ",color='k')
plt.scatter(x2,y2,label="low income high salary",color='m')

plt.show()
```

```python
grades= 'A','A-','B','B-','C','D','E'
gradeCount=[50,60,140,160,30,20,50]

plt.pie(gradeCount,labels=grades,autopct='%1.1f%%')

plt.show()
```

```python
import numpy as np

x=np.arange(0,5* np.pi,0.1)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_cos, 'g--')
plt.plot(x,y_sin, 'r--')
plt.title("Sine and Cosine")

plt.show()
```
